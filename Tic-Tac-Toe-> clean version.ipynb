```python
from IPython.display import clear_output

def player_input():
    while True:
        global marker
        marker = input('Please choose which marker you want, X or O: ')
        marker = marker.upper()
        if marker == 'X':
            return(marker,'You are X.')
        elif marker == 'O':
            return(marker,'You are O.')
        else: 
            print("you've make a wrong decision, try again")
            continue
        pass
    
def place_marker(board, marker, position):
    board[position-1]=marker
    return(board)
    


def display_board(board):
    print(board[6],board[7],board[8],sep='|')
    print('-'*5)
    print(board[3],board[4],board[5],sep='|')
    print('-'*5)
    print(board[0],board[1],board[2],sep='|')
    pass

def space_check(board, position):
    if board[position-1] == ' ':
        return True
    else:
        return False

def full_board_check(board):
    if ' ' in board.values():
        return True
    else:
        return False

def win_check(board):
    for marker in ['X', 'O']:
        if board[0]==board[1]==board[2]==marker:
            print ('{} wins'.format(marker))
            break
        elif board[0]==board[4]==board[8]==marker:
            print ('{} wins'.format(marker))
            break
        elif board[3]==board[4]==board[5]==marker:
            print ('{} wins'.format(marker))
            break
        elif board[6]==board[7]==board[8]==marker:
            print ('{} wins'.format(marker))
            break
        elif board[0]==board[3]==board[6]==marker:
            print ('{} wins'.format(marker))
            break
        elif board[1]==board[4]==board[7]==marker:
            print ('{} wins'.format(marker))
            break
        elif board[2]==board[5]==board[8]==marker:
            print ('{} wins'.format(marker))
            break
        elif board[2]==board[4]==board[6]==marker:
            print ('{} wins'.format(marker))
            break
        else:
            print("It's a tie.")
            break
    

from random import randint

def choose_first():
    global marker
    first = randint(1,2)
    if first == 1:
        print('X goes first.')
        marker = 'X'
        return marker
    else:
        print('O goes first.')
        marker = 'O'
        return marker
    pass            

def replay():
    global try_again #記得global statement
    response = input('Try again: ')
    if response == 'Yes':
        return True
    else:
        try_again = False
        return try_again

def player_choice(board):
    b = True
    while b:
        global position
        position = input('Please input a number from 1 to 9: ')
        lst = []
        for i in range(1,10):
            lst.append(str(i))
        if position in lst:
            position = int(position)
            if position <= 9:
                while space_check(board, position):
                    clear_output()
                    return(position)
            else:
                print('Wrong Nums, Stupid!')
                continue
        elif position not in lst:
            print('Please enter "Numbers" correctly, Stupid')
            continue
        print("It's been occupied. Try again!")
        continue
   

#####################################################################################################################

print('Welcome to Tic Tac Toe!')

try_again = True
while try_again:
    marker=''
    position = ' '
    board={}
    for i in range(0,9):
        board[i]=position
    player_input()
    choose_first()
    while full_board_check(board):
        player_choice(board)
        place_marker(board, marker, position)
        display_board(board)
        if marker == 'X':
            marker = 'O'
        else:
            marker = 'X'
    win_check(board)
    replay()
```
